```python
df1 = pd.get_dummies(df.Pclass).iloc[:,1:]
df1.head()
```

```python
df2 = pd.get_dummies(df.Sex).iloc[:,1:]
df2.head()
```

```python
df3=pd.get_dummies(df.Embarked).iloc[:,1:]
df3.head()
```

```python
df = pd.concat([df,df1,df2,df3],axis=1)
df.head()
```

```python
df.drop(['Pclass','Sex','Embarked'],axis=1,inplace=True)
```

```python
df.head()
```

```python
 x=df.iloc[:,1:]
```

```python
x.head()
```

```python
y=df.iloc[:,:1]
y.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
y.shape
```

```python
x.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accu = accuracy_score(y_test,y_pred)
accu
```

```python
from sklearn.naive_bayes import GaussianNB
model_naive = GaussianNB()
model_naive.fit(x_train,y_train)
```

```python
y_gauss_pred = model_naive.predict(x_test)
```

```python
accu_gauss = accuracy_score(y_test,y_gauss_pred)
accu_gauss
```
